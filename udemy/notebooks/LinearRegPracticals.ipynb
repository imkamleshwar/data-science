#### Creating Dummy Variables

```python
data_with_dummies = pd.get_dummies(data_no_mc, drop_first=True)
data_with_dummies.head()
```

```python
col = data_with_dummies.columns.values
```

```python
newCol = ['logPrice','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']

data_preprocessed = data_with_dummies[newCol]
data_preprocessed.head()
```

### Linear Regressions

```python
target = data_preprocessed['logPrice']
inputs = data_preprocessed.drop(['logPrice'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, target, test_size=0.2, random_state=365)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
yhat = lr.predict(x_train)
```

```python
plt.scatter(y_train, yhat)
plt.xlabel('Targets(y_train)', size=18)
plt.ylabel('Predictions(y_hat)', size=18)
plt.xlim(6, 13)
plt.ylim(6, 13)
plt.show()
```

##### Residual plot

```python
sns.displot(y_train - yhat, kde=True)
plt.title('Residuals PDF')
plt.show()
```

```python
lr.score(x_train, y_train)
```

### Weights and bias

```python
lr.intercept_
```

```python
lr.coef_
```

```python
lr_summary = pd.DataFrame(data=inputs.columns.values, columns=['Feature'])
lr_summary['Weights'] = lr.coef_
lr_summary
```

### Testing

```python
y_hat_test = lr.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets(y_test)', size=18)
plt.ylabel('Predictions(y_hat_test)', size=18)
plt.xlim(6, 13)
plt.ylim(6, 13)
plt.show()
```

```python
df_pf = pd.DataFrame(data=np.exp(y_hat_test), columns=['Predictions'])
df_pf
```

```python
df_pf['Targets'] = np.exp(y_test.reset_index(drop=True))
df_pf
```

```python
df_pf['Residual'] = df_pf['Targets'] - df_pf['Predictions']
df_pf
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual'] / df_pf['Targets'] * 100)
df_pf
```
